# silent letters in English

```python
import time
def time_stamp():
    result = time.localtime()
    # print("result:", result)
    # print("\nyear:", result.tm_year)
    # print("tm_hour:", result.tm_hour)
    named_tuple = time.localtime() # get struct_time
    # time_string = time.strftime("%m_%d_%Y_%H_%M_%S", named_tuple)
    time_string = time.strftime("%M_%S", named_tuple)
    # now_time = time.strftime("%m_%d_%Y_%H_%M_%S", named_tuple)
    return time_string
```

# silent B

```python
# pronunciation: silent letters

## pronunciation silent B Graphviz




from graphviz import Graph, Digraph
# see names and examples for different engines: https://nbviewer.jupyter.org/github/xflr6/graphviz/blob/master/examples/graphviz-engines.ipynb

g = Digraph('G', engine="circo")  

mydict = {"silent-B": ["debt",
                       "comb",
                       "bomb",
                       "doubt", ]}


key = "silent-B"
values = mydict["silent-B"]


for v in values:
    print (f"{v},")

with open("/content/sw_silent_b.txt","w") as myfile:
    for v in values:
        myfile.write(f"{v}\n")
# print("file is written")


for v in values:
    g.edge(key, v, "no 'b' sound")
    

#this will print the png but will _not_ display it
g.format = "png"  # this line changes the format to png ( svg is also possible, works in firefox browser )
# g.render() # this line saves the file as png
g.render(filename=f"/content/{time_stamp()}_{key}_graph") # this line saves the file as png

# have to put the g after the code!
g
```
